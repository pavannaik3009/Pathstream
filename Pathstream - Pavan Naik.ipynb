```python
ax = udemy.groupby("subject")["num_subscribers"].sum().sort_values(ascending = False).head(4)\
    .plot.bar(colors="#ffcc00", figsize=(12,8), fontsize=13, edgecolor='#008FFF')
for i in ax.patches:
    ax.annotate(np.round(i.get_height()), (i.get_x()+i.get_width()/2, i.get_height()),
         ha='center', va='center', color='#4F616B',fontsize =14,
         xytext=(2,8), textcoords='offset points')
ax.yaxis.grid()
ax.set_title('Number of students enrolled vs subjects (Udemy)', fontsize=16, color='black', pad=20)
ax.axhline(udemy.groupby("subject")["num_subscribers"].sum().min(), color='blue', linewidth=2, linestyle='--')
ax.axhline(udemy.groupby("subject")["num_subscribers"].sum().median(), color='red', linewidth=2, linestyle='-.')
plt.ylabel('Number of Subscribers',fontsize=14, color='black')
plt.xlabel('Subjects',fontsize=14, color='black')
plt.xticks(rotation=0)
ax.yaxis.labelpad = 20
ax.xaxis.labelpad = 20
plt.show()
```
